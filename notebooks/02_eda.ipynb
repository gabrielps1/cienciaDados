```python
# Célula 1 — imports e caminhos
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual compatível com Colab
plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline

# Caminho do dataset (upload manual)
PROCESSED_DIR = '/content'

# Pasta onde iremos salvar as figuras da EDA
REPORT_FIG_DIR = '/content/figuras_eda'
os.makedirs(REPORT_FIG_DIR, exist_ok=True)

print("Processed dir:", PROCESSED_DIR)
print("Figures dir:", REPORT_FIG_DIR)
```

```python
# Célula 2 — carregar dataset processado
fp = '/content/dataset_processado.csv'

df = pd.read_csv(fp, parse_dates=['timestamp'], low_memory=False)

print("Shape:", df.shape)
display(df.head(5))
display(df.dtypes)
```

```python
# Célula 3 — Missing, duplicados e estatísticas
print("Valores ausentes:")
display(df.isna().sum().sort_values(ascending=False).head(20))

print("\nLinhas duplicadas:", df.duplicated().sum())

print("\nResumo estatístico:")
display(df.describe(include='all').T)
```

```python
# Célula 4 — Criar colunas de data/hora
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['hour'] = df['timestamp'].dt.hour
df['date'] = df['timestamp'].dt.date
df['dayofweek'] = df['timestamp'].dt.dayofweek

display(df[['timestamp', 'hour', 'date', 'dayofweek']].head())
```

```python
# Célula 5 — Agregações úteis
agg = df.groupby(['region', 'hour']).agg(
    n=('timestamp','count'),
    mean_speed=('speed','mean') if 'speed' in df.columns else ('delay','mean'),
    median_speed=('speed','median') if 'speed' in df.columns else ('delay','median'),
    mean_delay=('delay','mean'),
    mean_rain=('rain_mm','mean'),
    mean_volume=('vehicle_volume','mean')
).reset_index()

display(agg.head())

agg.to_csv(os.path.join(REPORT_FIG_DIR, 'agg_region_hour.csv'), index=False)
```

```python
# Célula 6 — Histogramas
plt.figure(figsize=(12,5))
df['speed'].hist(bins=40)
plt.title("Distribuição da velocidade")
plt.xlabel("Velocidade (km/h)")
plt.ylabel("Frequência")
plt.savefig(os.path.join(REPORT_FIG_DIR, 'hist_speed.png'), dpi=150)
plt.show()

plt.figure(figsize=(12,5))
df['delay'].hist(bins=40)
plt.title("Distribuição do delay")
plt.xlabel("Delay (min)")
plt.ylabel("Frequência")
plt.savefig(os.path.join(REPORT_FIG_DIR, 'hist_delay.png'), dpi=150)
plt.show()
```

```python
# Célula 7 — Boxplot velocidade por região
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='region', y='speed')
plt.title("Boxplot de velocidade por região")
plt.savefig(os.path.join(REPORT_FIG_DIR, 'boxplot_speed_region.png'), dpi=150)
plt.show()
```

```python
# Célula 8 — Série temporal da velocidade média por dia
ts = df.groupby(['date','region'])['speed'].mean().unstack()

plt.figure(figsize=(14,5))
ts.plot()
plt.title("Velocidade média por dia")
plt.ylabel("Velocidade (km/h)")
plt.xlabel("Data")
plt.tight_layout()
plt.savefig(os.path.join(REPORT_FIG_DIR, 'ts_speed_region.png'), dpi=150)
plt.show()
```

```python
# Célula 9 — Heatmap de correlação
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlação entre variáveis numéricas")
plt.savefig(os.path.join(REPORT_FIG_DIR, 'heatmap_corr.png'), dpi=150)
plt.show()
```

```python
# Célula 10 — Média por hora
by_hour = df.groupby('hour')['speed'].mean()

plt.figure(figsize=(12,4))
plt.plot(by_hour.index, by_hour.values, marker='o')
plt.title("Velocidade média por hora")
plt.xlabel("Hora")
plt.ylabel("Velocidade média (km/h)")
plt.grid(True)
plt.savefig(os.path.join(REPORT_FIG_DIR, 'speed_by_hour.png'), dpi=150)
plt.show()

by_hour.head()
```

```python
# Célula 11 — Comparação chuva x seco
df['is_rain'] = df['rain_mm'] > 0.1

comp = df.groupby('is_rain')['speed'].mean()
display(comp)

comp.to_csv(os.path.join(REPORT_FIG_DIR, 'rain_vs_norain.csv'))
```

```python
# Célula 12 — Insights automáticos
insights = []

# 1. Horários de pico
top_pico = df.groupby('hour')['speed'].count().sort_values(ascending=False).head(3)
insights.append(("Horas com mais registros", top_pico.to_dict()))

# 2. Diferença média de velocidade com chuva
mean_rain = df[df['is_rain']==True]['speed'].mean()
mean_dry = df[df['is_rain']==False]['speed'].mean()
insights.append(("Diferença chuva/seco (km/h)", float(mean_dry - mean_rain)))

# 3. Região mais lenta
regiao_lenta = df.groupby('region')['speed'].mean().sort_values().head(1)
insights.append(("Região com menor velocidade média", regiao_lenta.to_dict()))

insights_df = pd.DataFrame(insights, columns=['Insight','Valor'])
display(insights_df)

insights_df.to_csv(os.path.join(REPORT_FIG_DIR, 'insights.csv'), index=False)
```
